```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('incendios-cantidad-causas-provincia_2022.csv',sep=';',encoding='latin-1')

df
```

```python
sns.barplot(df,x='incendio_anio',y='incendio_total_numero')

plt.xticks(rotation=70)

plt.show()
```

```python
df_2=pd.read_csv('superficie-incendiada-provincias-tipo-de-vegetacion_2022.csv',sep=';',encoding='latin-1')

df_2
```

```python
df_2['superficie_afectada_por_incendios_total_hectareas'] = df_2['superficie_afectada_por_incendios_total_hectareas'].apply(lambda x: pd.to_numeric(x.replace('-','0').replace()))
```
