```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('incendios-cantidad-causas-provincia_2022.csv',sep=';',encoding='latin-1')

df
```

```python
fig, ax= plt.subplots(figsize=(12,6))

plt.title('Cantidad total de incendios por año',size=14,weight='bold',pad=25)

sns.barplot(df,x='incendio_anio',y='incendio_total_numero',ci=None,color='darkblue')

ax.set_ylabel("Total de Incendios")
ax.set_xlabel(None)
ax.spines[['right', 'top']].set_visible(False)

plt.xticks(rotation=70)

for i in ax.containers:
    ax.bar_label(i,fontsize=8,weight='bold',rotation=65)

plt.show()
```

```python
df_2=pd.read_csv('superficie-incendiada-provincias-tipo-de-vegetacion_2022.csv',sep=';',encoding='latin-1')

df_2
```

```python
df_2['superficie_afectada_por_incendios_total_hectareas'].head(20)
```

```python
df_2['superficie_afectada_por_incendios_total_hectareas'] = df_2['superficie_afectada_por_incendios_total_hectareas'].str.replace('-','0').str.replace('.','').str.replace(',','.')


#df_2['superficie_afectada_por_incendios_total_hectareas']=pd.to_numeric(df_2['superficie_afectada_por_incendios_total_hectareas'])
```

```python
df_2['superficie_afectada_por_incendios_total_hectareas']= df_2['superficie_afectada_por_incendios_total_hectareas'].apply(lambda x: float(x))
```

```python
fig, ax= plt.subplots(figsize=(12,4))

plt.title('Cantidad de hectáreas \n quemadas por año',weight='bold',x=0.8,size=20)

sns.barplot(df_2,x='superficie_afectada_por_incendios_anio',y='superficie_afectada_por_incendios_total_hectareas',ci=None,color='darkblue')

ax.set_ylabel(None)
ax.set_xlabel(None)

ax.spines[['right', 'top']].set_visible(False)

plt.xticks(rotation = 70)

for i in ax.containers:
    ax.bar_label(i,fontsize=8,weight='bold',rotation=65)

plt.show()
```

```python
tierra_quemada=df_2.groupby(['superficie_afectada_por_incendios_provincia','superficie_afectada_por_incendios_anio']).mean().reset_index()
```

```python
tierra_quemada

```

```python
incendios_total=df_2.groupby(['superficie_afectada_por_incendios_anio'])['superficie_afectada_por_incendios_provincia','superficie_afectada_por_incendios_total_hectareas'].max().reset_index()

incendios_total.head()
```

```python
import matplotlib.ticker as ticker

fig, ax= plt.subplots(figsize=(12,4))

plt.title('Provincia con hectáreas \n más quemadas por año',weight='bold',x=0.8,size=20)

sns.barplot(incendios_total,x='superficie_afectada_por_incendios_anio',y='superficie_afectada_por_incendios_total_hectareas',color='darkblue')

ax.set_ylabel("Total de superficies incendiadas por hectáreas",weight='bold')
ax.set_xlabel(None)
ax.spines[['right', 'top']].set_visible(False)

plt.xticks(rotation = 70)


for i, v in enumerate(incendios_total['superficie_afectada_por_incendios_total_hectareas']):
    plt.text(i , v-1, incendios_total['superficie_afectada_por_incendios_provincia'][i],ha='center',va='bottom',rotation=90,weight='bold')

    
# Disable scientific notation on the y-axis
ax = plt.gca()
formatter = ticker.ScalarFormatter(useMathText=False)
formatter.set_scientific(False) 
ax.yaxis.set_major_formatter(formatter)

#plt.subplots_adjust(top=0.95)

plt.show()
```
